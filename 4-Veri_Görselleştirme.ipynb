# Dağılım BarPlot

```python
import seaborn as sns
import pandas as pd
diamonds=sns.load_dataset('diamonds')
df=diamonds.copy()
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.describe().T
```

```python
from pandas.api.types import CategoricalDtype
```

```python
df.cut=df.cut.astype(CategoricalDtype(ordered=True))
df.color=df.color.astype(CategoricalDtype(ordered=True))
df.clarity=df.clarity.astype(CategoricalDtype(ordered=True))
```

```python
df.cut.head(1) #Fair,Good, Very Good, Premium, İdeal
```

```python
df.color.head(1) #from j to D
```

```python
df.clarity.head(1) #from I1 to IF
```

```python
cut_kategoriler=['Fair','Good','Very Good','Premium','Ideal']
```

```python
df.cut=df.cut.astype(CategoricalDtype(categories=cut_kategoriler,ordered=True))
```

```python
df.cut.head(1)
```

```python
color_kategoriler=['J','I','H','G','F','E','D']
```

```python
df.color=df.color.astype(CategoricalDtype(categories=color_kategoriler,ordered=True))
```

```python
df.color.head(1)
```

```python
clarity_kategoriler=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
df.clarity=df.clarity.astype(CategoricalDtype(categories=clarity_kategoriler,ordered=True))
```

```python
df.clarity.head(1)
```

```python
from pandas_datareader import data
import pandas as pd
apple=data.DataReader('AAPL','yahoo',2014,2018)
df=apple.copy()
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn
```

```python
kapanis=df['Close']
kapanis.head()
```

```python
df.shape
```

```python
kapanis.plot();
```

```python
kapanis.index=pd.DatetimeIndex(kapanis.index)
```

```python
kapanis.plot();
```

```python
%config InlineBackend.figure_format='retina'
import seaborn as sns
sns.set_context('poster')
sns.set(rc={'figure.figsize':(16,9.)})
sns.set_style('whitegrid')
```

```python
kapanis.plot(style='-');
```

```python
# Hareketli ortalama bulunacak 
yedi=kapanis.rolling(window=7).mean()

on=kapanis.rolling(window=10).mean()

yirmi=kapanis.rolling(window=20).mean()

otuz=kapanis.rolling(window=30).mean()

yedi.head(10)
```

```python
kapanis.plot(style='-');
yedi.plot(style='--');
on.plot(style='-.');
yirmi.plot(style='.-');
otuz.plot(style='.');
plt.legend(['input','yedi','on','yirmi','otuz'], loc='lower right');
plt.xlim('2017','2018');
```
